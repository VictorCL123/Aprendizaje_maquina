### Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Lectura de datos

```python
df = pd.read_csv("data/Salario.csv")
X = df.iloc[:,:-1].values
y = df.iloc[:,1].values
```

### Dividir los datos

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=1/3,random_state=0)
```

### Regresión lineal

```python
regresion = LinearRegression()
regresion.fit(x_train,y_train)
```

```python
y_pred = regresion.predict(x_test)
```

### Visualización (entrenamiento)

```python
plt.scatter(x_train,y_train, color="red")
plt.plot(x_train,regresion.predict(x_train),color="blue")
plt.title("Sueldo y años de experiencia utilizando el conjunto de entrenamiento")
plt.xlabel("Años de experiencia")
plt.ylabel("Ingresos")
plt.show()
```

### Visualización (validación)

```python
plt.scatter(x_test,y_test, color="red")
plt.plot(x_train,regresion.predict(x_train),color="blue")
plt.title("Sueldo y años de experiencia utilizando el conjunto de validación")
plt.xlabel("Años de experiencia")
plt.ylabel("Ingresos")
plt.show()
```
