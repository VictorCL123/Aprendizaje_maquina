```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv("Salarios.csv")
```

```python
X = df.iloc[:,1:2].values
y = df.iloc[:,2].values
```

## Arbol

```python
regresion = DecisionTreeRegressor(random_state=0)
```

```python
regresion.fit(X,y)
```

```python
plt.scatter(X,y,color = 'green')
plt.plot(sorted(X),regresion.predict(sorted(X)),color='blue')
plt.title("Arbol")
plt.xlabel("Posicion del empleado")
plt.ylabel("Sueldo")
```

## Bosques

```python
regresion = RandomForestRegressor(random_state=0)
```

```python
regresion.fit(X,y)
```

```python
plt.scatter(X,y,color = 'green')
plt.plot(sorted(X),regresion.predict(sorted(X)),color='blue')
plt.title("Bosques")
plt.xlabel("Posicion del empleado")
plt.ylabel("Sueldo")
```
